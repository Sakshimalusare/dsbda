```python
import pandas as pd
from scipy.stats import zscore, skew
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\malus\Downloads\StudentsPerformance.csv")
```

```python
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
skew(df['math score'])
```

```python
df['zscore'] = (df['math score']-df['math score'].mean())/df['math score'].std()
df
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['math score'], kde=True, color='blue')
plt.title('Original Math Score')

plt.subplot(1, 2, 2)
sns.histplot(df['zscore'], kde=True, color='green')
plt.title('Z-score Transformed Math Score')

plt.tight_layout()
plt.show()
```
