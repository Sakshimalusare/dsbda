```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\malus\OneDrive\Desktop\dsbda\Iris.csv")
df.head()
```

```python
df.dtypes
```

```python
for column in df.columns:
    dtype = df[column].dtype
    if dtype == 'object':
        print(f"- {column}: Nominal")
    else:
        print(f"- {column}: Numeric")
```

```python
df.hist(figsize=(10, 6), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features')
```

```python
#12&13
plt.figure(figsize=(10, 6))
for i, column in enumerate(df.select_dtypes(include=['float']), 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[column], color='lightgreen')
    plt.title(f'Boxplot of {column}')
```

```python
# 12
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
numeric_features = df.select_dtypes(include=['float']).columns

for ax, feature in zip(axes.flat, numeric_features):
    sns.boxplot(y=df[feature], ax=ax, color='lightcoral')
    ax.set_title(f'Boxplot: {feature}')

plt.suptitle("Subplot Demonstration: Boxplots of Iris Features")
plt.tight_layout()
plt.show()
```

```python
# 13
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
numeric_cols = df.select_dtypes(include=['float']).columns

for ax, col in zip(axes.flat, numeric_cols):
    ax.hist(df[col], bins=15, color='orchid', edgecolor='black')
    ax.set_title(f'Histogram: {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.suptitle("Subplot Demo: Histograms of Iris Features", y=1.03)
plt.show()
```
