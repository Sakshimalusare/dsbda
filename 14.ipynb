```python
import pandas as pd

# Read the data
df = pd.read_csv(r"C:\Users\malus\OneDrive\Desktop\dsbda\weather.csv")
df 
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
```

```python
yearly_avg = df.groupby('Year')[['temp_max','temp_min', 'precipitation', 'wind']].mean()
```

```python
monthly_avg = df.groupby(['Year', 'Month'])[['temp_max','temp_min', 'precipitation', 'wind']].mean()
```

```python
print("Year-wise Averages:\n", yearly_avg)
print("\nMonth-wise Averages:\n", monthly_avg)
```
