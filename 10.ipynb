```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = sns.load_dataset("titanic")
df
```

```python
df.info()
```

```python
fare=df['fare'].dropna()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.hist(fare, bins=30, edgecolor='black')
plt.title('Fare Distribution - Histogram (matplotlib)')
```

```python
sns.distplot(df['fare'].dropna(), kde=True, color='orange')
plt.title('Distplot of Fare')
```

```python
sns.histplot(df['fare'].dropna(), kde=True, bins=30, color='purple')
plt.title('Histplot of Fare with KDE')
```

```python
sns.jointplot(x='fare', y='fare', data=df, kind='scatter', color='green')
plt.suptitle('Jointplot of Fare (Self vs Self)', y=1.02)
```

```python
sns.rugplot(df['fare'].dropna(), height=0.1, color='red')
plt.title('Rugplot of Fare')
```

```python
df['fare_bin'] = pd.cut(df['fare'], bins=[0, 50, 100, 150, 200, 250, 600])
sns.barplot(x='fare_bin', y='fare', data=df)
```
